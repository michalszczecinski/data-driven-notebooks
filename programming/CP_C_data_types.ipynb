#Computer Programming - Data Types

* C supports several different types of data, each of which may be represented differently within the computer's memory.
* In this notebook we list some of the basic types and use Python utility functions to demonstrate range of values.

```python
import pandas as pd
from IPython.display import Image
```

#### Utility functions

```python
def create_bit_table(bits_strings):
  """
  Produce table showing range of values given string of bits.
  """
  bits = bits_strings.replace(" ","")
  l = []
  for i,x in enumerate(bits):
    # x is bit, i is power starting from 0
    d = {'bit': x,
        'power': i,
        '2_to_power': 2**i,
        'position_value': int(x) * 2**i}
    l.append(d)
  df = pd.DataFrame(l)
  df['sum_value'] = df['position_value'].cumsum()
  return df.T

def convert_ascii_coded_message(ascii_message):
  """
  Take string of decimal values representing ASCII characters and
  return decoded string.
  """
  ascii_codes = ascii_message.split()
  # convert to in by applying int function to whole iterable
  ascii_codes = list(map(int, ascii_codes))
  # Printing input list
  print ("Initial list : ", ascii_codes)
  # convert decimal values to ASCII characters and join to one string
  decoded = ''.join(map(chr, ascii_codes))
  return decoded

def print_full_ascii_table():
  """
  Iterate through decimal values and correspinding ASCII characters.
  """
  digit = 0
  while digit<=255:
      print("%d = %c" %(digit, chr(digit)))
      digit += 1

def calculate_signed_range(no_bits):
  number_of_values = 2**no_bits
  signed_min_value = round((-1) * number_of_values/2)
  signed_max_value = number_of_values + signed_min_value - 1 # we deduct 1 because there is 0 in between
  print(f"Values from {signed_min_value} to {signed_max_value} \n")
  result = {'number_of_values': number_of_values,
       'signed_min_value': signed_min_value,
       'signed_max_value': signed_max_value}
  return result
```

## char
* char takes up 1 byte (8-bit) of memory that stores a character value
* character are written in single quotes, e.g. 'A'.
* Characters are stored as numbers as listed in ASCII table.

### What are ASCII Values?
ASCII stands for American Standard Code for Information Interchange. It is a character encoding that uses numeric codes to represent characters from 0 to 127. These include upper and lowercase English letters, numbers, and punctuation symbols. For example, the ASCII code for character A is 65, and Z is 90.

## byte
* byte stores an 8-bit unsigned numbers from 0 to 255
* 0000 0000 (0) to 1111 1111 (255)
* byte x = B100010; // 'B' in binary form

```python
bits = '1111 1111'
create_bit_table(bits)
```

```python
bits = '0111 0101'
create_bit_table(bits)
```

## word
* word stores 16-bit unsigned numbers from 0 to 65535
* same as unsigned integer
*  e.g. word = y = 10000;

```python
bits = '1111 1111 1111 1111'
create_bit_table(bits)
```

## int
* int is the data type for number storage
* stores 2-byte (16-bit) signed numbers 
* this data type has values from -32786 to +32767
* e.g. int ledPin = 15;

```python
bits = '1111 1111 1111 1111'
create_bit_table(bits)
```

```python
no_bits = 16
calculate_signed_range(no_bits)
```

## unsigned int
* unsigned int stores a 16-bit unsigned numbers from 0 to 65535
* e.g. `unsigned int count = 200;`

```python
bits = '1111 1111 1111 1111'
create_bit_table(bits)
```

## long
* long stores a 4-byte(32-bit) signed numbers from -2,147,483,648 to +2,147,483,647
* e.g. `long speedOfLight = 10000`;

```python
bits = '1111 1111 1111 1111 1111 1111 1111 1111'
create_bit_table(bits)
```

```python
no_bits = 32
calculate_signed_range(no_bits)
```

## unsigned long
* long stores a 4-byte(32-bit) unsigned numbers from 0 to +4,294,967,295
* e.g. `unsigned long numberOfVisitor = 2600000`;

## float
* float is the data type for number with decimal point. This data has values from 3.4028235E-38 to 3.4028235E+38.
* The data are stored as 4 bytes(32 bits) of information.
* e.g. `float sensorVal = 1.117`


## double

* double normally would be 8 bytes and higher precision than float with a greater range. However, on Arduino, it is the same as float.

**Note:** If data exceeds their range, then strange things happen. 
* If you have a byte of unsigned variable (range of 0 to 255), with value 255 and you add 1 to it, you get 0. 
* If you have an int variable equal to 23767 and you add 1 to it, you will end up with -32768.

### C variable - range inspection

To get the exact size of a type or a variable on a particular platform, you can use the `sizeof` operator. The expressions `sizeof(type`) yields the storage size of the object or type in bytes. Given below is an example to get the size of various type on a machine using different constant defined in `limits.h` header file. 

```python
%%writefile types.c
#include <stdio.h>
#include <stdlib.h>
#include <limits.h>
#include <float.h>

int main(int argc, char** argv) {
    
    printf("CHAR_BIT    :   %d\n", CHAR_BIT);
    printf("CHAR_MAX    :   %d\n", CHAR_MAX);
    printf("CHAR_MIN    :   %d\n", CHAR_MIN);
    printf("INT_MAX     :   %d\n", INT_MAX);
    printf("INT_MIN     :   %d\n", INT_MIN);
    printf("LONG_MAX    :   %ld\n", (long) LONG_MAX);
    printf("LONG_MIN    :   %ld\n", (long) LONG_MIN);
    printf("SCHAR_MAX   :   %d\n", SCHAR_MAX);
    printf("SCHAR_MIN   :   %d\n", SCHAR_MIN);
    printf("SHRT_MAX    :   %d\n", SHRT_MAX);
    printf("SHRT_MIN    :   %d\n", SHRT_MIN);
    printf("UCHAR_MAX   :   %d\n", UCHAR_MAX);
    printf("UINT_MAX    :   %u\n", (unsigned int) UINT_MAX);
    printf("ULONG_MAX   :   %lu\n", (unsigned long) ULONG_MAX);
    printf("USHRT_MAX   :   %d\n", (unsigned short) USHRT_MAX);

    printf("Storage size for float : %d \n", sizeof(float));
    printf("FLT_MAX     :   %g\n", (float) FLT_MAX);
    printf("FLT_MIN     :   %g\n", (float) FLT_MIN);
    printf("-FLT_MAX    :   %g\n", (float) -FLT_MAX);
    printf("-FLT_MIN    :   %g\n", (float) -FLT_MIN);
    printf("DBL_MAX     :   %g\n", (double) DBL_MAX);
    printf("DBL_MIN     :   %g\n", (double) DBL_MIN);
    printf("-DBL_MAX     :  %g\n", (double) -DBL_MAX);
    printf("Precision value: %d\n", FLT_DIG );

    return 0;
}
```

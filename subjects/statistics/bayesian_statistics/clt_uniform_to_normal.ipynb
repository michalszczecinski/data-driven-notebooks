# Central Limit Theorem - uniform to normal distribution

1. This notebook is about the central limit theorem. It shows how the uniform distribution can be transformed into a normal distribution by taking the average of many samples from the uniform distribution.

```python
# imports
import numpy as np
import pandas as pd
import math
import random
import unittest

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('darkgrid')
```

### Functions

```python
def generate_samples(x, number_of_samples, sample_size):
    """
    Generate a list of samples from a population.

    Args:
        x: list or array-like object representing the population from which the samples are drawn
        number_of_samples: integer representing the number of samples to generate
        sample_size: integer representing the size of each sample

    Returns:
        List of lists, where each sublist is a sample drawn from the population x.

    """
    all_samples = []
    for i in range(number_of_samples):
        sample = random.sample(x, sample_size)
        all_samples.append(sample)
    return all_samples
```

```python
def show_sample_statistics(all_samples):
    """
    Calculate the mean and sum of each sample in a list of samples, and plot a histogram of the means and sums.

    Args:
        all_samples: list of lists, where each sublist is a sample drawn from a population

    Returns:
        None.

    """
    averages = []
    sums = []
    for sample in all_samples:
        avg = np.array(sample).mean()
        sum = np.array(sample).sum()
        averages.append(avg)
        sums.append(sum)
    # Assume averages and sums are two lists of data to plot
    fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
    axs[0].hist(averages, color='blue')
    axs[0].set_title('Averages')
    axs[1].hist(sums, color='orange')
    axs[1].set_title('Sums')
    plt.show()
```

## Problem: Show that samples from uniform distribution form normal distribution.

Example 1: Guessing a Birthday

If you walked up to a random person on the street, the probability that their birthday falls on a given date would follow a uniform distribution because each day of the year is equally likely to be their birthday.

For example, there are 365 days in a year so the probability that their birthday is on January 1st would be 1/365.
Similarly, the probability that their birthday is on January 2nd is 1/365. And so on.

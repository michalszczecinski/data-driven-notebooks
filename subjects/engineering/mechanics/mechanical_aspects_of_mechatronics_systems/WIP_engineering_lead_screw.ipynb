## Translational Motion Converters

```python
display(Image('images/lead_screw_input_output.png'))
```

For one full circle rotation the translation is 1 pitch (distance between the grooves). Imagine the ant walking up the thread for full circle.
For translational motion converter the output is the straighline movement, the input is one full circle (2pi radians), so the velocity ration is:
$$
n = \frac{output}{input} = \frac{p}{2\pi} \text{, where p is value of pitch}
$$

## Generate multiple triangles

### Configuration

```
r=1.0
x1 = 0
y1=0
samples = 11
```

```
# we can draw radius lines using 
# pythagorean theorem and example x coordinates

x2arr = np.linspace(0,1, num = samples)
df = pd.DataFrame()
df['r'] = [r for x in range(1, samples+1)]
df['x1'] = x1
df['y1']=y1

df['x2'] = np.linspace(0,1, num = samples)
df['y2'] = (df['r']**2 - df['x2']**2)**0.5

df['cos_a'] = df['y2'] / r
df['sin_a'] = df['x2'] / r
df['a_rad'] = np.arcsin(df['sin_a'])
df['a_degrees'] =df['a_rad'] * (180/math.pi)

df
```

```
plt.axes()
circle = plt.Circle((0,0),1, facecolor='None',edgecolor='blue')
plt.gca().add_patch(circle)

n = 11
color=iter(cm.rainbow(np.linspace(0,1,n)))

for index, row in df.iterrows():
  c=next(color)
  print(index)
  print(row)
  print(c)
  line1 = plt.Line2D((row['x1'], row['x2']), (row['y1'], (row['y2'])), linewidth=1.5, linestyle='--', color = c)
  line2 = plt.Line2D((row['x2'], row['x2']), (row['y2'], (row['y1'])), linewidth=1.5, linestyle='--', color = c)
  plt.gca().add_line(line1)
  plt.gca().add_line(line2)
  

plt.axis('scaled')
plt.show()
```

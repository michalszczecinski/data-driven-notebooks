## OPEN-LOOP CONTROL

- Open-loop systems do not compare the actual result with the desired result to determine the control action.
- there is no feedback signal or measuring device.
- To achieve control, the system is given a calibrated setting.
- the control action is fixed unless the set point is changed or the whole system is re-calibrated to another working condition.
- in the event of load changes or disturbances, there is no way for the system to know that the controlled variable has been affected and hence there is no compensation in the control action to bring the controlled variable back to setpoint.

Advantages
- generally less expensive than the closed-loop
- not necessary to measure the actual result
- the controller is simple in design
- no instability problem


Disadvantages

- errors caused by unexpected disturbances and load changes are not corrected
- recalibration is required from time to time

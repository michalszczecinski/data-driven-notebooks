## Radians and Degrees

image.png

* The measure of an angle is determined by the amount of rotation from the initial side to the terminal side.
* In radians, one complete counterclockwise revolution is 2 and in degrees, one complete counterclockwise revolution is 360.

So, degree measure and radian measure are related by the equations

$$
\boxed{
360= 2\pi \text{ radians} \\
180= \pi \text{ radians}
}
$$

From the latter, we obtain the equation

$$
\boxed{
1 \text{ radian} = \frac{360}{2 \pi} \text{ degrees} \\
1 \text{ radian} = \frac{180}{\pi} degrees \\
1 \text{ rad} \approx 57.296 
}
$$
<br>
$$
\boxed{
1 \text{ degree} = \frac{\pi}{180} \text{ radians} \\
1 = 0.018 \text{ rad}
}
$$

* To convert from radians to degrees, multiply the radians by $\frac{180}{}$
* To convert from degrees to radians, multiply the degrees by $\frac{\pi}{180} \text{ rad}$

##  Examples

### Example 1

Convert 20 to radians.


$$
360 = 2 \pi \text{ |divide both sides by 18} \\
20 = \frac{2 \pi}{18} = \frac{\pi}{9} \text{ rad}
$$ 

### Example 2
Convert $\frac{4 \pi}{3}$ rad to degrees.

$$
\frac{4 \pi}{3} \text{ rad}=\frac{4 \pi}{3} * \frac{180}{\pi}=240 \text{ |pi cancels, do not forget the degree symbol}
$$
